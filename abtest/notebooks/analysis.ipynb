### Calculando resultados por grupo

```python

query_resposta_1 = """
SELECT
    c.grupo,
    COUNT(DISTINCT c.cliente_id) AS cont_clientes,
    COUNT(DISTINCT p.pedido_id) AS cont_pedidos,
    SUM(p.valor_total) AS valor_pedidos,
    SUM(p.valor_total)/(COUNT(DISTINCT c.cliente_id)) AS ticket_medio,
    ROUND(1.0 * COUNT(DISTINCT p.pedido_id)/COUNT(DISTINCT p.cliente_id),2) as taxa_pedidos
FROM
df_clientes c
INNER JOIN df_pedidos p ON c.cliente_id = p.cliente_id
WHERE
p.status = "Confirmado"
GROUP BY c.grupo
"""
resposta_1 = sqldf(query_resposta_1, globals())
print(resposta_1)
```

### Calculando aumento real dos grupos Variantes por Controle - Taxa de Pedidos

```python
taxa_A = resposta_1.loc[resposta_1['grupo'] == 'A', 'taxa_pedidos'].values[0]

resposta_1['aumento_real_taxa_pedidos'] = (resposta_1['taxa_pedidos'] - taxa_A) / taxa_A

resposta_1['aumento_real_taxa_pedidos%'] = resposta_1['aumento_real_taxa_pedidos'] * 100

print(resposta_1[['grupo', 'taxa_pedidos', 'aumento_real_taxa_pedidos%']])
```

**Criação de gráficos**

```python
import matplotlib.pyplot as plt

resposta_1_ordenado = resposta_1.sort_values(by='grupo', ascending=False)
grupos = resposta_1_ordenado['grupo'].tolist()
vals = resposta_1_ordenado['aumento_real_taxa_pedidos%'].tolist()
colors = ['lightblue' if v > 0 else ('lightgrey' if v < 0 else 'lightgrey') for v in vals]

plt.figure(figsize=(10,3))
plt.barh(grupos, vals, color=colors)
plt.axvline(0, color='black', linewidth=1)
for i, v in enumerate(vals):
    plt.text(v + (0.15 if v>=0 else -0.15), i, f'{v:.1f}%',
             va='center', ha='left' if v>=0 else 'right', fontsize=9)
plt.xlabel('Aumento real da taxa de pedidos (%)')
plt.xlim(-100, 100)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

### Calculando aumento real dos grupos Variantes por Controle - Ticket Médio

```python
tm_a = resposta_1.loc[resposta_1['grupo'] == 'A', 'ticket_medio'].values[0]

resposta_1['aumento_real_tm'] = (resposta_1['ticket_medio'] - tm_a) / tm_a

resposta_1['aumento_real_tm%'] = resposta_1['aumento_real_tm'] * 100

print(resposta_1[['grupo', 'ticket_medio', 'aumento_real_tm%']])
```

### Calculando taxa de confirmação por grupo

```python
query_resposta_1_1 = """
SELECT
    t.grupo,
    (t.cont_pedidos - t.pendentes) as pedidos_finalizados,
    t.confirmados,
    round(100.0 * t.confirmados / NULLIF(t.cont_pedidos - t.pendentes, 0), 2) AS taxa_confirmacao
FROM(
    SELECT
        c.grupo,
        COUNT(DISTINCT c.cliente_id) AS cont_clientes,
        COUNT(DISTINCT p.pedido_id) AS cont_pedidos,
        COUNT(DISTINCT CASE WHEN p.status = 'Confirmado' THEN p.pedido_id END) AS confirmados,
        COUNT(DISTINCT CASE WHEN p.status = 'Pending' THEN p.pedido_id END) AS pendentes,
        COUNT(DISTINCT CASE WHEN p.status = 'CANCELADO' THEN p.pedido_id END) AS cancelados
    FROM
    df_clientes c
    INNER JOIN df_pedidos p ON c.cliente_id = p.cliente_id
    GROUP BY c.grupo) t
GROUP BY t.grupo
"""
resposta_1_1 = sqldf(query_resposta_1_1, globals())
print(resposta_1_1)
```

```python
import matplotlib.pyplot as plt

resposta_1_1_ordenado = resposta_1_1.sort_values(by='grupo', ascending=False)
grupos = resposta_1_1_ordenado['grupo'].tolist()
vals = resposta_1_1_ordenado['taxa_confirmacao'].tolist()
colors = ['lightblue' if v > 80 else 'lightgrey' for v in vals]

plt.figure(figsize=(8,5))
plt.bar(grupos, vals, color=colors)
for i, v in enumerate(vals):
    plt.text(i, v + 0.3, f'{v:.2f}%',
             ha='center', va='bottom', fontsize=9)
plt.ylabel('Taxa de Confirmação (%)')
plt.ylim(0, 100)
plt.xlabel('')
plt.tight_layout()
plt.show()
```

Subquery para determinar faixas de qtde de pedidos
Query para calcular a taxa de confirmação por faixa de pedidos

### Criando as faixas de qtde de pedidos baseado na distribuição dos dados

```python
query_resposta_4_2 = """
SELECT
    t.faixa,
    COUNT(DISTINCT t.pedido_id) as pedidos_totais,
    COUNT(DISTINCT CASE WHEN t.status = 'Confirmado' then t.pedido_id end) as pedidos_confirmados,
    ROUND(
      100.0 * COUNT(DISTINCT CASE WHEN t.status = 'Confirmado' THEN t.pedido_id END)
      / NULLIF(COUNT(DISTINCT t.pedido_id), 0),
      2
    ) AS taxa_confirmacao
    FROM (
        SELECT
        i.pedido_id,
        SUM(i.quantidade) as qtde,
        CASE WHEN sum(i.quantidade) >= 10 THEN '10+'
        WHEN sum(i.quantidade) BETWEEN 8 AND 9 THEN '8-9'
        WHEN sum(i.quantidade) BETWEEN 6 AND 7 THEN '6-7'
        WHEN sum(i.quantidade) BETWEEN 4 AND 5 THEN '4-5'
        WHEN sum(i.quantidade) <=3 THEN '<=3'
        END as faixa,
        p.status
        FROM
        df_itenspedido i
        INNER JOIN df_pedidos p 
        ON i.pedido_id = p.pedido_id
        INNER JOIN df_produtos pr
        ON i.produto_id = pr.produto_id
        group by i.pedido_id, p.status) t
WHERE t.status IN ('Confirmado','CANCELADO')
GROUP BY t.faixa
ORDER BY qtde DESC
"""

resposta_4_2 = sqldf(query_resposta_4_2, globals())
print(resposta_4_2)
```

```python
import matplotlib.pyplot as plt

resposta_4_2_sorted = resposta_4_2.sort_values('taxa_confirmacao', ascending=True)

estados = resposta_4_2_sorted['faixa'].tolist()
vals = resposta_4_2_sorted['taxa_confirmacao'].tolist()

media_taxa = sum(vals) / len(vals)

plt.figure(figsize=(8, 3))
plt.barh(estados, vals, color='lightblue')
plt.axvline(media_taxa, color='red', linewidth=2, linestyle='--', alpha=0.4)

for i, v in enumerate(vals):
    plt.text(v + max(vals)*0.01, i, f'{v:,}'.replace(',', '.'), va='center', fontsize=9)

plt.xlabel('Taxa de Confirmação (%)')
plt.ylabel('Itens por Pedido')
plt.xlim(0, max(vals) * 1.2)
plt.legend()
plt.tight_layout()
plt.show()
```

### Houve algum aumento na quantidade de itens pedidos por cliente na nova funcionalidade?

```python
query_resposta_5 = """
SELECT
  a.categoria,
  ROUND(a.pedidos_por_cliente, 3) AS ppc_a,
  ROUND(c.pedidos_por_cliente, 3) AS ppc_c,
  ROUND(100.0 * (c.pedidos_por_cliente / NULLIF(a.pedidos_por_cliente, 0) - 1), 2) AS variacao_c_a
FROM (
  SELECT
    pr.categoria,
    1.0 * COUNT(DISTINCT CASE WHEN p.status IN ('Confirmado','CANCELADO') THEN p.pedido_id END)
      / NULLIF(COUNT(DISTINCT CASE WHEN p.status IN ('Confirmado','CANCELADO') THEN p.cliente_id END), 0)
      AS pedidos_por_cliente
  FROM df_pedidos p
  JOIN df_itenspedido i ON i.pedido_id = p.pedido_id
  JOIN df_produtos pr ON pr.produto_id = i.produto_id
  JOIN df_clientes c ON c.cliente_id = p.cliente_id
  WHERE c.grupo = 'A' AND p.status IN ('Confirmado','CANCELADO')
  GROUP BY pr.categoria
) a
JOIN (
  SELECT
    pr.categoria,
    1.0 * COUNT(DISTINCT CASE WHEN p.status IN ('Confirmado','CANCELADO') THEN p.pedido_id END)
      / NULLIF(COUNT(DISTINCT CASE WHEN p.status IN ('Confirmado','CANCELADO') THEN p.cliente_id END), 0)
      AS pedidos_por_cliente
  FROM df_pedidos p
  JOIN df_itenspedido i ON i.pedido_id = p.pedido_id
  JOIN df_produtos pr ON pr.produto_id = i.produto_id
  JOIN df_clientes c ON c.cliente_id = p.cliente_id
  WHERE c.grupo = 'C' AND p.status IN ('Confirmado','CANCELADO')
  GROUP BY pr.categoria
) c
  ON a.categoria = c.categoria
ORDER BY variacao_c_a DESC;
"""

resposta_5 = sqldf(query_resposta_5, globals())
print(resposta_5)
```

```python
import matplotlib.pyplot as plt

resposta_5_ordenado = resposta_5.sort_values(by='variacao_c_a', ascending=True)
estados = resposta_5_ordenado['categoria'].tolist()
vals = resposta_5_ordenado['variacao_c_a'].tolist()
colors = ['lightblue' if v > 0 else ('lightgrey' if v < 0 else 'lightgrey') for v in vals]

plt.figure(figsize=(10,3))
plt.barh(estados, vals, color=colors)
plt.axvline(0, color='black', linewidth=1)
for i, v in enumerate(vals):
    plt.text(v + (0.15 if v>=0 else -0.15), i, f'{v:.1f}%',
             va='center', ha='left' if v>=0 else 'right', fontsize=9)
plt.xlabel('Variação do Nº de Itens por Pedido de Cliente por Categoria (%)')
plt.xlim(-100, 100)
plt.ylabel('')
plt.tight_layout()
plt.show()
```
